```python
import pandas as pd
import requests
url = "https://api.spacexdata.com/v4/launches"
```

```python
response = requests.get(url)
df = pd.DataFrame(response.json())
```

```python
df.head(5)
```

```python
df[['name', 'date_utc', "success", "cores", "payloads", "rocket", "launchpad"]].head(12)
```

```python
rockets = requests.get("https://api.spacexdata.com/v4/rockets").json()
rocket_dict = {r['id']: r['name'] for r in rockets}
df['rocket_name'] = df['rocket'].map(rocket_dict)
```

## here we extract launchpad's name info, and their coordinates, and proceed to store them in columns

```python
launchpads = requests.get("https://api.spacexdata.com/v4/launchpads").json()

launchpads_dict = {l['id']: l['name'] for l in launchpads}
coordinates_dict = {l['id']: {'latitude': l['latitude'], 'longitude': l['longitude']} for l in launchpads}

df['launchpad_name'] = df['launchpad'].map(launchpads_dict)
df['launch_lat'] = df['launchpad'].map(lambda x: coordinates_dict[x]['latitude'])
df['launch_long'] = df['launchpad'].map(lambda x: coordinates_dict[x]['longitude'])
```

```python
df[['rocket_name', 'launchpad_name', 'launch_lat', 'launch_long']].tail(12)
```

```python
df.columns
```

## Extracting fields from the cores section in our dataset

```python
df['cores'].head()
```

```python
# checking if the core was reused
df['core_reused'] = df['cores'].apply(lambda x: x[0]['reused'] if x and isinstance(x,list) and x[0] else None)

# checking if landing was a success
df['landing_success'] = df['cores'].apply( lambda x: x[0]['landing_success'] if x and isinstance(x, list) and x[0] else None)

# type of landing
df['landing_type'] = df['cores'].apply(lambda x: x[0]['landing_type'] if x and isinstance(x, list) and x[0] else None)

# checking if gridfins were used or not
df['gridfins_usage'] = df['cores'].apply(lambda x: x[0]['gridfins'] if x and isinstance(x,list) and x[0] else None)

# number of flights of the core
df['numberof_flights'] = df['cores'].apply(lambda x: x[0]['flight'] if x and isinstance(x,list) and x[0] else None)

df[['core_reused', 'landing_success', 'landing_type', 'gridfins_usage', 'numberof_flights']].head(12)
```

# checking for missing values in all finalised columns

```python
df.isnull().sum()
```

```python
df['numberof_flights'].fillna(0, inplace=True)
df['gridfins_usage'].fillna(False, inplace=True)
df['landing_success'].fillna('Unknown', inplace=True)
df['landing_type'].fillna('Unknown', inplace=True)
df['success'].fillna('Unknown', inplace=True)
df['details'].fillna('Unknown', inplace=True)
df['static_fire_date_utc'].fillna(0, inplace=True)
df['static_fire_date_unix'].fillna(0, inplace=True)
df['window'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

## Kaplan-Meier Survival
 how likely is a booster to still be functioning after N number of flights

```python
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt
```

```python
kmf = KaplanMeierFitter()
```

```python
df[['flight_number', 'landing_success']].dtypes
```

```python
df['flight_number'] = pd.to_numeric(df['flight_number'], errors='coerce')
df['landing_success'] = pd.to_numeric(df['landing_success'], errors='coerce')
df_survival = df[['flight_number', 'landing_success']].dropna()
```

```python
kmf.fit(durations=df_survival['flight_number'], event_observed=df_survival['landing_success'])

plt.figure(figsize=(8,5))
kmf.plot()
plt.title("kaplan-meier curve")
plt.xlabel("reuse count")
plt.ylabel("probability of survival")
plt.grid(True)
plt.show()
```

## Geospatial visualization with Folium

```python
import folium
from folium.plugins import MarkerCluster
```

```python
launch_map = folium.Map(location=[28.5, -80.5], zoom_start=3)
cluster = MarkerCluster().add_to(launch_map)
```
